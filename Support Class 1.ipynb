## Simple Random Walk

**Defintion Simple Random Walk** - Let $x_1, ... \in \{-1,1\}$ be iid r.v with $\mathbb{P}[x_i = 1] = p, \mathbb{P}[x_i = -1] = q$. Then the sequence $Y_0, Y_1,...$ defined as $Y_0 = 0$ and $Y_n = \sum_{k=1}^{n} x_k$ is the simple random walk on $\mathbb{Z}$

```python
#parameters 
p = 0.5
q = 1 - p
t = 0
tmax = 100
sum_x = 1 #initial starting place
y_0 = 0
time_plot=[t]
Y = [y_0]
%matplotlib notebook

start_time = time.time()

fig = plt.figure(figsize = (10,4))
ax = fig.add_subplot(111)

plt.ion
plt.title('Simple Random Walk')

fig.show()
fig.canvas.draw()

while t <tmax:
    r = np.random.rand()
    if r<p:
        x = 1
    else:
        x = -1
    sum_x = sum_x + x
    
    ax.clear()
    ax.plot(time_plot, Y)
    t +=1
    Y.append(sum_x)
    time_plot.append(t)
    fig.canvas.draw()
    plt.xlabel('Time')
    plt.ylabel('state')
    plt.tight_layout()
    
print("--- %s seconds ---" % (time.time() - start_time))
```

Recall,

$\mathbb{E}[Y_n] = \mathbb{E}\left[ \sum_{k=1}^{n} x_k \right] = \sum_{k=1}^{n} = n(2p-1)$ 

and 

$ var[Y_n] = var\left[ \sum_{k=1}^{n} x_k \right] = \sum_{k=1}^{n} var[x_k] = 4np(1-p)$

We will run N simulations of a random walk and calculate numerically the average and deviations over the replications and compare this to the theoretical solution.

```python
# Run simple random walk 10 times
# X is the random walk variables
# Y is the cumulative distance from origin, sum of X_n

def SRW(p, tmax, N):
    q = 1-p
    X = np.random.rand(N,tmax)
    X[X<p] = 1 # Convert all cells less than p to 1
    X[X!=1] = -1 # Convert all other cells to -1
    Y = np.zeros((N,tmax))
    for i in range(0, N):
        # For each random walk, bulid Y
        X[i][0] = 0 #start at zero - should probably be Y[i][0]
        Y[i,:] = np.cumsum(X[i,:]) # add up x's up to n for each y
    return X, Y

Xs, Ys = SRW(0.3, 10, 5)

print("Xs: ", Xs)
print("\nYs: ", Ys)

# Unclear, X[X<p] = 1. Variable used within index test is same variabel anem as used for actual array - confusing and unnecessary
```
